```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris=datasets.load_iris()
x=iris.data
fig=plt.figure(figsize=(12,12))
ax=Axes3D(fig)
X=PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X[:,0],X[:,1],X[:,2],c=iris.target,s=100)
ax.set_title("Iris Data with 1st 3 PCA directions")
ax.set_xlabel("1st eigen vector")
ax.set_ylabel("2nd eigen vector")
ax.set_zlabel("3rd eigen vector")
plt.show()
```
